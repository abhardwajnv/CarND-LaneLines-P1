## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

# Not Edited
def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Not Edited
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

# Not Edited
def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

# Not Edited
def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

# New Function Added
def fit_line(lines, y1, y2):
    """
    This Function averages and extrapolates the line segments using fitline approach.
    
    It takes input of y co-ordinates of predominent line and array of all co-ordinates from detected edges
    and provides an output in the form on final co-ordinates of the line.
    """
    x_array = []
    y_array = []
    for line in lines:
        xi1, yi1, xi2, yi2 = line[0]
        x_array.append(xi1)
        x_array.append(xi2)
        y_array.append(yi1)
        y_array.append(yi2)

    slope, intercept = np.polyfit(x_array, y_array, 1)  # Compute the slope and intercept of the line.

    x1 = int((y1 - intercept) / slope)
    x2 = int((y2 - intercept) / slope)
    return x1, y1, x2, y2

# Edited
def draw_lines(img, lines, color=[255, 0, 0], thickness=12):
    
    """ # Enable this Function to see all detected lines without further processing.
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), [0,0,255], 2)
    """
    
    left_array = []
    right_array = []
    imshape = img.shape
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = 0
            if x2 - x1 != 0.0:   # Make sure (x2-x1) is not Zero to avoid "Divide by Zero" Error
                slope = (y2 - y1) / (x2 - x1)
            if slope > 0 and x1 > (imshape[1] / 2) and x2 > (imshape[1] / 2):   # Right Lines Detected
                right_array.append(line)        # Append all the right line coordinates to right_array
            elif slope < 0 and x1 < (imshape[1] / 2) and x2 < (imshape[1] / 2):  # Left Lines Detected
                left_array.append(line)         # Append all the left line coordinates to left_array

    left_x1, left_y1, left_x2, left_y2 = fit_line(left_array, int(imshape[0]), int(imshape[0] / 2 + 55))
    right_x1, right_y1, right_x2, right_y2 = fit_line(right_array, int(imshape[0]), int(imshape[0] / 2 + 55))
    
    # Draw the lines on the edges from Canny Detection
    cv2.line(img, (left_x1, left_y1), (left_x2, left_y2), color, thickness)
    cv2.line(img, (right_x1, right_y1), (right_x2, right_y2), color, thickness)

# Not Edited
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.
# Not Edited
def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline


````python
def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)
    
    # Apply Grayscale Transform on the image and then Gaussian Noise Blur on Grayscale Image
    gray = grayscale(image)
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)

    # Apply Canny Edge Detection Algorithm to detect edges
    # by identifying pixels with strongest gradients from blur_gray image.
    low_threshold = 20
    high_threshold = 150
    edges = canny(blur_gray, low_threshold, high_threshold)

    # Region of Interest masked on the canny edges images
    # by defining a polygon and filling the edges inside that region.
    imshape = image.shape
    vertices = np.array([[
                (50, imshape[0]),#(0,imshape[0]),
                (imshape[1]/2 - 50,3*imshape[0]/5),#(470, 310),
                (imshape[1]/2 + 50,3*imshape[0]/5),#(490, 310),
                (imshape[1]-50,imshape[0]),#(imshape[1],imshape[0])
            ]], dtype=np.int32)
    masked_image = region_of_interest(edges, vertices)

    # Apply Hough Transform on the Canny Edges to identify lines from the edges and 
    # then processing them to identify the most prodominent line and filling that on the edges.
    rho = 1
    theta = np.pi/180
    threshold = 1
    min_line_length = 10
    max_line_gap = 5
    line_image = hough_lines(masked_image, rho, theta, threshold, min_line_length, max_line_gap)
    result = weighted_img(line_image, image)    
    return result
````
